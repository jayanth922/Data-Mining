```python
import time
import numpy as np
import pandas as pd

from pycaret.datasets import get_data
from pycaret.time_series import TSForecastingExperiment
```

```python
data = pd.read_csv('/content/AirPassengers.csv')
data['Month'] = pd.to_datetime(data['Month'])
data = data.set_index('Month')
```

```python
forecast_horizon = 12
folds = 3

fig_settings = {
    "renderer": "png",
    "width": 1000,
    "height": 600,
}
```

```python
experiment = TSForecastingExperiment()
experiment.setup(data=data, target='#Passengers', fh=forecast_horizon, fig_kwargs=fig_settings)
```

```python
experiment.plot_model()
```

```python
experiment.plot_model(data_kwargs={"plot_data_type": ["original", "imputed", "transformed"]})
```

```python
experiment.plot_model(plot="acf")
```

```python
experiment.plot_model(plot="pacf", data_kwargs={'nlags':36}, fig_kwargs={'height': 500, "width": 800})
```

```python
experiment.plot_model(plot="periodogram")
```

```python
experiment.plot_model(plot="fft")
```

```python
experiment.plot_model(plot="diagnostics", fig_kwargs={"height": 800, "width": 1000})
```

```python
experiment.plot_model(
    plot="diff",
    data_kwargs={"lags_list": [[1], [1, 12]], "acf": True, "pacf": True, "periodogram": True},
    fig_kwargs={"height": 800, "width": 1500}
)
```

```python
experiment.plot_model(plot="decomp", fig_kwargs={"height": 500})
```

```python
experiment.plot_model(plot="decomp", data_kwargs={'seasonal_period': 24}, fig_kwargs={"height": 500})
```

```python
experiment.plot_model(plot="decomp", data_kwargs={'type': 'multiplicative'}, fig_kwargs={"height": 500})
```

```python
experiment.plot_model(plot="decomp_stl", fig_kwargs={"height": 500})
```

```python
experiment.plot_model(plot="train_test_split", fig_kwargs={"height": 400, "width": 900})
```

```python
experiment.plot_model(plot="cv", fig_kwargs={"height": 400, "width": 900})
```
